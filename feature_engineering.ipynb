```python
import pandas as pd
from pathlib import Path

df = pd.read_csv("data/cleaned/superstore_cleaned.csv")

# Ensure date format
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")

# Add year and month
df["order_year"] = df["order_date"].dt.year
df["order_month"] = df["order_date"].dt.month

# High value order flag
df["high_value_order"] = (df["sales"] > 500).astype(int)

# Sales category bucket
df["sales_bucket"] = pd.cut(
    df["sales"],
    bins=[0, 100, 500, 1000, df["sales"].max()],
    labels=["Low", "Medium", "High", "Very High"]
)

# Create curated folder if not exists
Path("data/curated").mkdir(parents=True, exist_ok=True)

# Save curated data
df.to_csv("data/curated/superstore_curated.csv", index=False)

print("Feature Engineering Completed Successfully")
```
