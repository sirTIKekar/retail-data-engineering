```python
import pandas as pd

df = pd.read_csv("data/cleaned/superstore_cleaned.csv")

# Convert date column
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")

print("Null Values:\n", df.isnull().sum())

# Check negative sales
if (df["sales"] < 0).any():
    print("Warning: Negative sales present")
else:
    print("No negative sales found")

# Date range check
print("Min Date:", df["order_date"].min())
print("Max Date:", df["order_date"].max())

# Hard validation
assert df["sales"].dtype != "object", "Sales column wrong type"
assert df["order_date"].isnull().sum() == 0, "Date conversion failed"

print("Data Validation Completed Successfully")
```
