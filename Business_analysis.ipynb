```python
import pandas as pd
import matplotlib.pyplot as plt

fact = pd.read_csv("warehouse/fact_sales.csv")

# Total Sales
print("Total Revenue:", fact["sales"].sum())

# Monthly Sales
monthly = fact.groupby("order_date")["sales"].sum()
monthly.plot(figsize=(12,5))
plt.title("Daily Sales Trend")
plt.show()

# Top Customers
top_customers = fact.groupby("customer_id")["sales"].sum().sort_values(ascending=False).head(10)
print(top_customers)
```
