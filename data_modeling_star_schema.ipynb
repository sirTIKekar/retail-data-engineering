```python
import pandas as pd
from pathlib import Path

df = pd.read_csv("data/curated/superstore_curated.csv")

Path("warehouse").mkdir(parents=True, exist_ok=True)

dim_customer = df[
    ["customer_id", "customer_name", "segment"]
].drop_duplicates()

dim_product = df[
    ["product_id", "product_name", "category", "sub-category"]
].drop_duplicates()

dim_location = df[
    ["country", "state", "city", "region"]
].drop_duplicates()

dim_date = df[
    ["order_date", "order_year", "order_month"]
].drop_duplicates()

fact_sales = df[
    [
        "order_id",
        "customer_id",
        "product_id",
        "order_date",
        "sales",
        "high_value_order"
    ]
]

# Save all tables
dim_customer.to_csv("warehouse/dim_customer.csv", index=False)
dim_product.to_csv("warehouse/dim_product.csv", index=False)
dim_location.to_csv("warehouse/dim_location.csv", index=False)
dim_date.to_csv("warehouse/dim_date.csv", index=False)
fact_sales.to_csv("warehouse/fact_sales.csv", index=False)

print("Star Schema Created Successfully")
```
